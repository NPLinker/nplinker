```python
import numpy
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
from nplinker.scoring import misc
```

```python
strains = 100

exp_map = numpy.zeros((strains, strains))
var_map = numpy.zeros((strains, strains))

for i in range(strains):
    for j in range(strains):
        exp, var = misc.metcalf_expected_count(i, j, None, strains, 10, -10, 0, -1)
        exp_map[i, j] = exp
        var_map[i, j] = var
```

```python

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4))

ax1.set_title('Expected value')
im1 = ax1.imshow(exp_map)

divider = make_axes_locatable(ax1)
cax = divider.append_axes("right", size="5%", pad=0.05)

fig.colorbar(im1, cax=cax)

ax1.set_xlabel('MF size')
ax1.set_ylabel('GCF size')



ax2.set_title('Variance')
im2 = ax2.imshow(var_map)

divider = make_axes_locatable(ax2)
cax = divider.append_axes("right", size="5%", pad=0.05)

fig.colorbar(im2, cax=cax)

ax2.set_xlabel('MF size')


fig.tight_layout()

plt.savefig('graph_correlation_expected_variance.png')
```
